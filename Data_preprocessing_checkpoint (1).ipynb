Hadling missing values is based on some critical points
1. What type of data is it, is it categorical or nnumerical
2. Will not having that whole column that appears to have more than 30% null values affect our data
3. Now, if they are important rows and columns, and numerical, i prefer to fill with the mean of that column
4. Also, if they categorical, mode of the column will suffice
5. If a column has more than 30% null values, it will be dropped
6, Ultimately, filling null values should be after in-depth understanding of the data
